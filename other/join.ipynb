```python
import pandas as pd

ls = [
    pd.read_csv(f'data/0408/data/validate_dataset_chunk_{i}.csv')
    for i in range(5)
]

df = pd.concat(ls, axis=0)
# df.to_csv('data/0408/data/validate_dataset_with_rw.csv', index=False)

df.shape

orig_df = pd.read_csv('data/0408/data/diverse_dataset.csv')

# df = pd.read_csv('data/0408/data/validate_dataset_with_rw.csv')
```

```python
test_df = df[['rewrite_prompt', 'original_text', 'rewritten_text']]

# Filter rows where 'rewritten_text' contains only ASCII characters
test_df = test_df[test_df['rewritten_text'].apply(lambda x: x.isascii() \
                                                  and '[Your Name]' not in x \
                                                  and '[Insert Date]' not in x)]

test_df['rewritten_text'] = test_df['rewritten_text'] \
                                .str.replace(r"^Sure.+\:", "", regex=True) \
                                .str.replace(r"^>", "", regex=True) \
                                .str.replace("Sure, here is the rewrite in limerick form:", "") \
                                .str.strip() \
                                .apply(lambda x: x if not (x.startswith('"""') and x.endswith('"""')) else x[3:-3])

print(test_df.shape)
# clean unneeded text


# first_sentence = test_df['rewritten_text'].str.split('.').str[0].str[:100]
# first_sentence


# [Your Name]

merged_df = pd.merge(test_df, orig_df, on=['rewrite_prompt', 'original_text'])
merged_df
```

```python
merged_df.to_csv('data/0408/data/data_final.csv', index=False)
```
